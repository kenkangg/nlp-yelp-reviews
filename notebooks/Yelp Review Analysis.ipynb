## Discussion
While we were unable to recieve results from our baseline model, we can compare our CNN and BERT experiements. There was a substantial improvment when using BERT. This was consistent with our expectations. Our assumption was built on the BERT's advantages over CNNs for sequence models:
    * Attention
    * Direcctly optimized word embeddings
    * Transfer Learning (Difficult to find pretrained models for CNN within NLP context)
    
Fortunately, we were also comparing our results to the rest of our class through a leaderboard, which is shown below. The image only shows 15 of the 33 teams involved and includes results based on a series of challenge datasets created by the course staff to test various edge cases. Our BERT model performed the best based on mean absolute error(MAE), but was outperformed by other groups when comparing accuracy.

This distinction was most apparent in our performance on the 6th challenge dataset, which consisted of only 1 and 5 star reviews. As seen below, we had the best MAE of the group, but the worst accuracy by a large margin. We hypothesize that this is a result of our choice in using a regression loss over classification. This would make it much harder to misclassify a 1-star review as a 5-star and vice-versa, but would also blur the line between a 1-star and 2-star review. As an example, our model initally outputs a continuous value such as 1.51 which would round to 2 stars. Even though our raw prediction implies uncertainty between ratings, we need a discrete output, so we lose this information and assume the closest answer. Our low accuracy score in challenge 6 was likely because of many samples misclassified as 2-star reviews.

## Future Work
Looking forward, there are experiements and improvements that could be done to further improve our model:
* While we did account for the imbalance of the data by undersampling the original dataset, we did not consider incorporationg the underlying bias post-prediction. One way this could have been done is by optimizing thresholds for rounding continuous values. This way if 1-star reviews are objectively more prevalent than 2-star reviews, we could make sure to only predict 2-star reviews when the prediction is > 1.6 as an example. 


* Additionally, another direction we could take is experimenting with classifcation models, or even regression-classification ensembles. This has the potential of tackling the issue mentioned in the discussion section.
